```python
import pandas as pd

dataset = pd.read_csv('/content/DW_Headphones3 - Sheet1.csv')
dataset
```

```python
dataset.dropna(inplace = True)
dataset
```

```python
dataset["4. Headphone usage"].fillna(4, inplace = True)
dataset
```

```python
dataset.loc[0, '11. Ear Cuffs'] = "Yes"
dataset["11. Ear Cuffs"]
```

```python
def update_ear_cuffs(value):
    if value != "Yes" and value != "No":
        return "No"
    return value

dataset["11. Ear Cuffs"] = dataset["11. Ear Cuffs"].apply(update_ear_cuffs)

print(dataset["11. Ear Cuffs"])
```
